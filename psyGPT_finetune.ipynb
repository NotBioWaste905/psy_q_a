```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
tok = GPT2Tokenizer.from_pretrained("models/us")
```

```python
model = GPT2LMHeadModel.from_pretrained("models/us")
```

```python
model.cuda()
```

```python
text = """PROMPT: как жить в таком жестоком мире? | в таком жестоком мире жить всегда тяжело, но нужно
ANSWER: """
inpt = tok.encode(text, return_tensors="pt")
```

```python
out = model.generate(inpt.cuda(), max_length=150, repetition_penalty=5.0, do_sample=True, top_k=5, top_p=0.95, temperature=1.5)
```

```python
print(tok.decode(out[0]))
```
